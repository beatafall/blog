```python
import pandas as pd, numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

Setup plot params

```python
def get_country(c,h='ds',plot=False):
    if c=='huro':
        hu=pd.read_csv('data/'+'hu'+'_'+h+'.csv') #daily data
        ro=pd.read_csv('data/'+'ro'+'_'+h+'.csv') #daily data
        df=pd.concat([hu,ro])
    else:
        df=pd.read_csv('data/'+c+'_'+h+'.csv') #daily data
        # df=pd.read_csv('data/'+c+'_hs.csv') #high_res data
    df['time']=pd.to_datetime(df['time'])
    df['year']=df['time'].dt.year
    df['month']=df['time'].dt.month
    df['day']=df['time'].dt.day
    df['hour']=df['time'].dt.hour
    df=df.set_index('time')
    df=df.sort_index()
    if plot: df.groupby('year').nunique()['ID'].plot()
    return df
```

```python
c='ro'
df=get_country(c,plot=True)
```

```python
keys=['XTEMP','YPCP','YSNW','XVSB']
dz=df.groupby(['ID','year']).mean()[keys]
dzm=dz.groupby(['ID']).mean()
dzn=dz-dzm
dzn['NT']=np.abs(stats.zscore(dzn['XTEMP']))
```
