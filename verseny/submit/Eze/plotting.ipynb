```python
data=pd.DataFrame.from_dict(json.loads(open("first.json").read()))
data2=pd.DataFrame.from_dict(json.loads(open("data.json").read()))
```

```python
def replace_place(string):
    to_replace=['balan', 'barot', 'csikszereda', 'gyergyo', 'keresztur', 'kezdi',
       'kolozsvar', 'kovaszna', 'parajd', 'regen', 'segesvar', 'szentgyorgy',
       'szovata', 'toplita', 'udvarhely', 'vasarhely', 'vlahica']
    replacement=['Bălan', 'Baraolt', 'Miercurea Ciuc', 'Gheorgheni', 'Cristuru Secuiesc', 'Târgu Secuiesc',
       'Cluj-Napoca', 'Covasna', 'Praid', 'Reghin', 'Sighisoara', 'Sfântu Gheorghe',
       'Sovata', 'Toplița, Harghita', 'Odorheiu Secuiesc', 'Târgu Mureș', 'Vlahița']
    for i in arange(len(to_replace)):
        if string==to_replace[i]: return replacement[i]
    return string
```

```python
def give_coordinates(string):
    p_lat=[]
    p_lon=[]
    located_p=[]
    for place in string:
        location = do_geocode(place)
        if(location):
            p_lat.append(location.latitude)
            p_lon.append(location.longitude)
            located_p.append(location)
        else: 
            p_lat.append(0)
            p_lon.append(0)
            located_p.append(None)
    return p_lat, p_lon, located_p
```

```python
data2["birth"]=data2["birth"].apply(replace_place)
data2
```

```python
x=data2.groupby("birth").count()
first_map=x[["id"]]
```

```python
p_lat,p_lon,located_p=give_coordinates(first_map.index)
located_p
```

```python
source = ColumnDataSource(data=dict(
    x = p_lon,
    y = p_lat,
    radius = first_map["id"]/20000,
    num = first_map["id"],
    name=list(first_map.index.values),
))

x_range=(min(p_lon)-(max(p_lon)-min(p_lon))/10,max(p_lon)+(max(p_lon)-min(p_lon))/10)
y_range=(min(p_lat)-(max(p_lat)-min(p_lat))/10,max(p_lat)+(max(p_lat)-min(p_lat))/10)

hover = HoverTool(tooltips=[
    ("Name", "@name"),
    ("Emberek száma", "@num"),],point_policy="follow_mouse")
p = figure(toolbar_location=None,   x_axis_location=None, y_axis_location=None,tools=[hover],
                                    x_range=x_range,y_range=y_range)
p.grid.grid_line_color = None

p.scatter("x","y",radius="radius",fill_alpha=0.6,
          line_color=None, source = source)

output_file("plotting.html", title="color_scatter.py example")
show(p)  # open a browser
```

```python
p_lat3=[]
p_lon3=[]
from pyproj import Proj, transform
for i in arange(len(p_lat2)):
    print(p_lat2[i],p_lon2[i])
    jj,ii=transform(Proj(init='epsg:4326'), Proj(init='epsg:3857'), p_lon2[i],p_lat2[i])
    p_lat3.append(ii)
    p_lon3.append(jj)
    #print(p_lat2[i],p_lon2[i])
```

```python
from bokeh.tile_providers import CARTODBPOSITRON, STAMEN_TONER
```

```python
source = ColumnDataSource(data=dict(
    x = p_lon3,
    y = p_lat3,
    radius = log(second_map["id"])*30000,
    num = second_map["id"],
    name=list(second_map.index.values),
))



x_range=-1000000,4000000
y_range=4000000,8500000

TOOLS="pan,wheel_zoom,reset,hover,save"
hover = HoverTool(tooltips=[
    ("Name", "@name"),
    ("Emberek száma", "@num"),],point_policy="follow_mouse")

p = figure( x_axis_location=None, y_axis_location=None,tools=[hover],
                                    x_range=x_range,y_range=y_range)
p.grid.grid_line_color = None
p.add_tools(WheelZoomTool())
p.add_tools(PanTool())

p.scatter("x","y",radius="radius",fill_alpha=0.6,
          line_color="black", source = source)

output_file("plotting2.html", title="color_scatter.py example")

show(p)
```
