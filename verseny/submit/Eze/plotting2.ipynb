```python
%pylab inline
import pandas as pd
import json

pd.options.display.max_rows = 2000

from IPython.core.display import HTML

from bokeh.io import output_notebook, show
output_notebook()
from bokeh.plotting import figure, show, output_file

from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    WheelZoomTool,
    PanTool,
)
from bokeh.tile_providers import CARTODBPOSITRON, STAMEN_TONER
```

```python
data = pd.read_pickle('data.pkl')
data2 = pd.read_pickle('data2.pkl')
```

```python
first_map=data2.groupby(["birth_located","birth_p_lat","birth_p_lon"],as_index=False).count()
```

```python
x = first_map["birth_p_lon"]
y = first_map["birth_p_lat"]

source = ColumnDataSource(data=dict(
    x = x,
    y = y,
    radius = first_map["id"].values/20000,
    num = first_map["id"].values,
    name=list(first_map["birth_located"]),
))

x_range=(min(x)-(max(x)-min(x))/10,max(x)+(max(x)-min(x))/10)
y_range=(min(y)-(max(y)-min(y))/10,max(y)+(max(y)-min(y))/10)

hover = HoverTool(tooltips=[
    ("Name", "@name"),
    ("Emberek száma", "@num"),],point_policy="follow_mouse")
p = figure(toolbar_location=None,   x_axis_location=None, y_axis_location=None,tools=[hover],
                                    x_range=x_range,y_range=y_range)#, border_fill_color=None)
p.grid.grid_line_color = None

p.scatter("x","y",radius="radius",fill_alpha=0.6,
          line_color=None, source = source)

output_file("newplot.html", title="color_scatter.py example")
show(p)  # open a browser
```

```python
second_map=data.groupby(["livesin_located","livesin_p_lat","livesin_p_lon"],as_index=False).count()
```

```python
p_lat3=[]
p_lon3=[]
from pyproj import Proj, transform
for i in arange(len(second_map["livesin_p_lon"])):
    #print(second_map["livesin_p_lon"][i],second_map["livesin_p_lat"][i])
    jj,ii=transform(Proj(init='epsg:4326'), Proj(init='epsg:3857'), second_map["livesin_p_lon"][i],second_map["livesin_p_lat"][i])
    p_lat3.append(ii)
    p_lon3.append(jj)
    #print(p_lat2[i],p_lon2[i])
```

```python
source = ColumnDataSource(data=dict(
    x = p_lon3,
    y = p_lat3,
    radius = log(second_map["id"])*5000,
    num = second_map["id"],
    name=list(second_map["birth_located"]),
))



x_range=-1000000,4000000
y_range=4000000,8500000

TOOLS="pan,wheel_zoom,reset,hover,save"
hover = HoverTool(tooltips=[
    ("Name", "@name"),
    ("Emberek száma", "@num"),],point_policy="follow_mouse")

p = figure( x_axis_location=None, y_axis_location=None,tools=[hover],
                                    x_range=x_range,y_range=y_range)
p.grid.grid_line_color = None
p.add_tools(WheelZoomTool())
p.add_tools(PanTool())

p.scatter("x","y",radius="radius",fill_alpha=0.6,
          line_color="black", source = source)

output_file("newplot2.html", title="color_scatter.py example")

show(p)
```
